# PINN Training Visualization
This notebook demonstrates how to load a PINN model, perform inferences, and visualize tracking uncertainty.

```python
import torch
import numpy as np
from models.pinn_module import PINNModule
from visualization.uncertainty_viz import plot_uncertainty
from simulation.trajectory_gen import TrajectoryGenerator
from preprocessing.blackout_labeler import BlackoutLabeler
```

```python
# Initialize a dummy PINN model
pinn = PINNModule(state_dim=6, hidden_dim=256, n_layers=6)
pinn.eval()
```

```python
# Generate mock data
gen = TrajectoryGenerator()
traj = gen.generate_batch(1)[0]
true_traj = traj['trajectory']
mask = BlackoutLabeler().label_trajectory(traj['plasma_profile'])
time_steps = np.arange(len(mask)) * 0.1
```

```python
# Simulate tracking errors and uncertainty covariance
N = len(true_traj)
errors = np.random.randn(N, 3) * 5.0
covs = np.zeros((N, 3, 3))
for i in range(N):
    if mask[i]:
        covs[i] = np.eye(3) * 100.0  # High uncertainty during blackout
    else:
        covs[i] = np.eye(3) * 25.0   # Normal radar uncertainty
```

```python
# Plot Uncertainty
plot_uncertainty(time_steps, errors, covs)
```
