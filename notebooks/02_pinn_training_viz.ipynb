# 02 - PINN Training and Visualization
Visualizing the Physics-Informed Neural Network training process and residuals.

```python
from physnet_hgv.models.pinn_module import PINNModule
import torch
import matplotlib.pyplot as plt

model = PINNModule()
t = torch.linspace(0, 10, 100).reshape(-1, 1)
out = model(t).detach().numpy()
plt.plot(t, out[:, 0], label='State X')
plt.legend()
plt.show()
```
