# Plasma Sheath Visualization
This notebook demonstrates the visualization of the HGV trajectory and the corresponding plasma sheath electron density profile.

```python
import numpy as np
import matplotlib.pyplot as plt
from simulation.trajectory_gen import TrajectoryGenerator
from preprocessing.blackout_labeler import BlackoutLabeler
from visualization.trajectory_viz import plot_trajectory_3d
from visualization.blackout_viz import plot_blackout_profile
from visualization.radar_viz import plot_radar_returns
```

```python
# Generate a sample trajectory
gen = TrajectoryGenerator()
traj = gen.generate_batch(1)[0]
```

```python
# Label blackout regions
labeler = BlackoutLabeler(blackout_threshold=1e18)
mask = labeler.label_trajectory(traj['plasma_profile'])
time_steps = np.arange(len(mask)) * 0.1
```

```python
# Plot Plasma Profile
plot_blackout_profile(time_steps, traj['plasma_profile'], threshold=1e18)
```

```python
# Plot Radar Returns
plot_radar_returns(time_steps, traj['trajectory'], traj['radar_returns'], mask)
```

```python
# Plot 3D Trajectory
plot_trajectory_3d(traj['trajectory'], blackout_mask=mask)
```
