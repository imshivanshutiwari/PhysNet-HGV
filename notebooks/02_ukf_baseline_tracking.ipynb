# Tracking Results
This notebook demonstrates the results of the full tracking pipeline using the Unscented Kalman Filter and PINN.

```python
import numpy as np
from filters.ukf_tracker import UKFTracker
from simulation.trajectory_gen import TrajectoryGenerator
from preprocessing.blackout_labeler import BlackoutLabeler
from evaluation.metrics import position_rmse, velocity_rmse
from visualization.trajectory_viz import plot_trajectory_3d
```

```python
# Generate a sample trajectory with blackout
gen = TrajectoryGenerator()
traj = gen.generate_batch(1)[0]
true_traj = traj['trajectory']
meas_traj = traj['radar_returns']
mask = BlackoutLabeler().label_trajectory(traj['plasma_profile'])
```

```python
# Initialize UKF Tracker
dt = 0.1
tracker = UKFTracker(dt=dt)
```

```python
# Run Tracker
estimates, covariances = tracker.run_filter(meas_traj, mask, pinn_model=None)
```

```python
# Evaluate RMSE
rmse_pos = position_rmse(estimates, true_traj)
rmse_vel = velocity_rmse(estimates, true_traj)
print(f"Position RMSE: {rmse_pos:.2f} m")
print(f"Velocity RMSE: {rmse_vel:.2f} m/s")
```

```python
# Plot Estimated vs True Trajectory
plot_trajectory_3d(true_traj, estimated_trajectory=estimates, blackout_mask=mask)
```
